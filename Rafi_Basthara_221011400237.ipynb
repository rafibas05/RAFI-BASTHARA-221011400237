```python
##############1##############
def sort_company(companies_dict):
    sorted_companies = sorted(companies_dict.values(), key=lambda x: x['domain'], reverse=True)
    result = [{"name": company['name'], "domain": company['domain']} for company in sorted_companies]
    return result

sorted_companies = sort_company(company_detail_list)
print(sorted_companies)
```

```python
##############2#############
def get_domain_company(companies_dict):
    for company_name, company_info in companies_dict.items():
        domain = company_info['domain']
        num_clients = len(company_info.get('clients', []))
        print(f"{company_name}: {domain}, relation: {num_clients} clients")

get_domain_company(company_detail_list)
print(get_domain_company)
```

```python
##############3##############
def get_employee(employee_detail_list, company_detail_list):
    employees_with_domain = []
    for emp_name, emp_info in employee_detail_list.items():
        company_name = emp_info['company']
        company_info = company_detail_list.get(company_name)
        if company_info:
            domain = company_info['domain']
            employees_with_domain.append({
                "full_name": emp_info['full_name'],
                "company": company_name,
                "domain": domain
            })
    return employees_with_domain

result = get_employee(employee_detail_list, company_detail_list)
print(result)
```

```python
#############4#############
def companies_with_employees(employee_detail_list, company_detail_list):
    companies_with_employees = {}
    # Initialize an empty list of employees for each company
    for company_info in company_detail_list.values():
        companies_with_employees[company_info['name']] = []

    # Populate the list of employees for each company
    for emp_info in employee_detail_list.values():
        company_name = emp_info['company']
        companies_with_employees[company_name].append(emp_info['full_name'])

    # Create the final output format
    result = [{"company": company, "employees": employees} for company, employees in companies_with_employees.items()]
    return result

result = companies_with_employees(employee_detail_list, company_detail_list)
print(result)
```

```python
from sklearn.preprocessing import StandardScaler

# Assuming your DataFrame is named df
empat = data_startup
scaler = StandardScaler()

# Iterate over each column and scale its values
for column in empat.columns:
    # Check if the column contains numerical data
    if empat[column].dtype in ['int64', 'float64']:
        # Reshape the column values to a 2D array (required by StandardScaler)
        column_values = empat[column].values.reshape(-1, 1)
        # Scale the column values
        scaled_values = scaler.fit_transform(column_values)
        # Update the column in the DataFrame with the scaled values
        empat[column] = scaled_values.flatten()

print(empat)
```
